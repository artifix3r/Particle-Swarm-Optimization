```python
#our particle class
class Particles():
    def __init__(own): #for generate particles 
        x = (-1) ** bool(rd.getrandbits(1)) * rd.random() * 1000
        y = (-1) ** bool(rd.getrandbits(1)) * rd.random() * 1000
        own.pos = np.array([x, y])
        own.pBest_value = float('inf')
        own.pBest_pos = own.pos
        own.velo = np.array([0,0])
    #for update function
    def update(own):
        own.pos = own.pos + own.velo
```

```python
class our_Space():
    def __init__(own, trg, trg_err, number_particles):
        own.trg = trg
        own.trg_err = trg_err
        own.number_particles = number_particles
        own.particles = []
        own.gBest_value = float('inf')
        own.gBest_pos = np.array([rd.random() * 50, rd.random() * 50])
            
    def fit(own, particles):
        x = particles.pos[0]
        y = particles.pos[1]
        f =  x**2 + y**2 + 1
        return f
    
    def set_pBest(own):
        for particle in own.particles:
            fit_candidate = own.fit(particle)
            if(particle.pBest_value > fit_candidate):
                particle.pBest_value = fit_candidate
                particle.pBest_pos= particle.pos
                
    def set_gBest(own):
        for particle in own.particles:
            best_fit_candidate = own.fit(particle)
            if(own.gBest_value > best_fit_candidate):
                own.gBest_value = best_fit_candidate
                own.gBest_pos = particle.pos
                
    def update_particles(own):
        for particle in own.particles:
            global W
            inertial = W * particle.velo
            own_confidence = c1 * rd.random() * (particle.pBest_pos - particle.pos)
            swarm_confidence = c2 * rd.random() * (own.gBest_pos - particle.pos)
            updated_velo = inertial + own_confidence + swarm_confidence
            particle.velo = updated_velo
            particle.update()
            
    def show_particles(own, iteration):        
        print(iteration, 'iterations')
        print('Bes tPosition in this time:', own.gBest_pos)
        print('Best Value in this time:', own.gBest_value)
        
        for particle in own.particles:
            plt.plot(particle.pos[0], particle.pos[1], 'ro')
        plt.plot(own.gBest_pos[0], own.gBest_pos[1], 'bo')
        plt.show()
```
